```python
# %matplotlib notebook
import os
import pandas as pd 
import math
import numpy as np 
import matplotlib.pyplot as plt
import random as r 
import glob
```

```python
nBodies = 3

# get sim files
workDir = "/Users/brandonmanley/Documents/nBody/data/pred/"

# Import data
file = workDir + "comb10_1_{0}.csv".format(nBodies)
df = pd.read_csv(file, index_col=False)
```

```python
df.head()
```

```python
# plot an individual event
###############################################

# pick event to plot
for i in range(5):
    r.seed(1)
    indexArray = np.array(df['eventID'].astype(int))
    indexArray = np.unique((indexArray/10000).astype(int))
    indexInt = np.random.choice(indexArray)
    print(indexInt)
    index = indexInt*10000
    # print(indexArray)

    # plot event 
    scale = 8
    edf = df.loc[(df['eventID'] >= index) & (df['eventID'] <= (index+2560))]
    fig, ax = plt.subplots()
    plt.rcParams["figure.figsize"] = (15*scale,10*scale)

    tsize = 60
    mark_density = 35
    linew = 40
    colors = ['red', 'blue', 'green', 'orange', 'purple']
    for i in range(1, nBodies+1):
        ax.plot('xf{0}'.format(i), 'yf{0}'.format(i), "", data=edf, marker='o',linewidth=linew, markersize=tsize, markevery=mark_density, color=colors[i-1])
        ax.plot('x{0}'.format(i), 'y{0}'.format(i),"", data=edf, marker='o',color='black', linewidth=linew, markersize=tsize*3)
#         ax.plot('xf{0}p'.format(i), 'yf{0}p'.format(i),"", data=edf, marker='o', linestyle=':', linewidth=linew, markersize=tsize, markevery=mark_density, color=colors[i-1])

    # plt.legend(loc='best', ncol=2, fancybox=True)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Event {0}'.format(int(index/10000)))
    # plt.show()
    plt.savefig('sample_NOPREDevent{0}_{1}body.png'.format(int(index/10000),nBodies))
    print("hi")
```
